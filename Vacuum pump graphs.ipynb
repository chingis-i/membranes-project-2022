```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '/content/drive/MyDrive/Colab Notebooks/Membranes project/Optim-results.xlsx'
results = pd.read_excel(path,sheet_name='Sheet1')
```

```python
results['C_vp'] = results['C_vp']*2
```

```python
results
```

```python
fig,axes = plt.subplots(2,3,figsize=(18,7),dpi=150)

graph1 = sns.lineplot(x='p_l',y='phi',marker='o',data=results,ax=axes[0,0])
axes[0,0].set(xlabel='Permeate side pressure, $p_l$ (mmHg)', ylabel='Stage-cut $\phi$')

graph2 = sns.lineplot(x='p_l',y='A',marker='o',data=results,ax=axes[0,1])
axes[0,1].set(xlabel='Permeate side pressure, $p_l$ (mmHg)', ylabel='Total membrane area $A_t$ ($cm^2$)')

graph3 = sns.lineplot(x='p_l',y='O2_rec',marker='o',data=results,ax=axes[0,2])
axes[0,2].set(xlabel='Permeate side pressure, $p_l$ (mmHg)', ylabel='O2 recovery, $R_{O_2}$ (%)')

graph4 = sns.lineplot(x='p_l',y='C_m',marker='o',data=results,ax=axes[1,0])
axes[1,0].set(xlabel='Permeate side pressure, $p_l$ (mmHg)', ylabel='Membrane cost, $C_m$ (\$)')

graph5 = sns.lineplot(x='p_l',y='C_vp',marker='o',data=results,ax=axes[1,1])
axes[1,1].set(xlabel='Permeate side pressure, $p_l$ (mmHg)', ylabel='Vacuum pump cost, $2 \cdot C_vp$ (\$)')

graph6 = sns.lineplot(x='p_l',y='C_c',marker='o',data=results,ax=axes[1,2])
axes[1,2].set(xlabel='Permeate side pressure, $p_l$ (mmHg)', ylabel='Total capital cost, $C_c$ (\$)')

plt.savefig('vacuum-pump-results.png')
plt.show()
```
